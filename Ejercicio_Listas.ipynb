```python
products = {"Pan": 10.25, "Crema": 55.25, "Mayonesa": 18.62, "Amaranto": 35,
            "Frijol": 32, "Aguacate": 15.22, "Vino tinto": 45.26, "Jamón": 26.22}

print("Carrito de compras\n")

in_car = []
show_car = False

while True:
    print(f"Nuestros Productos:\n")
    for item in products:
        print(f"{item}{'-' * (50 - (len(str(float(products[item]))) + len(item)))}${float(products[item])}")

    product = input("""
Ingresa el producto que necesitas o digita S para ir directo al carrito
""").capitalize()

    if product in products:
        amount = int(input("\n¿Cuanto quieres comprar?\n"))

        if amount >= 1:

            price = products[product]

            in_car.append([amount, product, price])  # Se crea una lista de listas

    elif product.lower() == "s":
        show_car = True
        if not in_car:
            print("Vuelva pronto")
            break

    else:
        print("Producto incorrecto")

    if show_car:
        print("Tu carrito de compras\n")
        total = 0
        for item in in_car:
            new_price = f"{(item[0] * item[2]):.2f}"  # Se calcula el precio final

            element = f"{item[0]} X {item[1]}${new_price}"

            total += float(new_price)

            print(f"{item[0]} X {item[1]}{'-' * (50 - len(element))}${new_price}")

        print(f"\nTOTAL{'-' * (50 - (len('TOTAL$') + len(f'{total:.2f}')))}${total:.2f}")

        break
```
