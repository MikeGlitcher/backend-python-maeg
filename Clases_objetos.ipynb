```python
products = {"Pan": 10.25, "Crema": 55.25, "Mayonesa": 18.62, "Amaranto": 35,
            "Frijol": 32, "Aguacate": 15.22, "Vino tinto": 45.26, "Jamón": 26.22}


class AddCar:
    car_elements = 0

    inventory = []

    temp_price = 0

    def add_ticket(self):
        self.inventory.append([amount, product, price])

        self.only_print()

    def only_print(self):
        for item in self.inventory:
            self.find_price(item)

            new_price = self.temp_price

            element = f"{item[0]} X {item[1]}${new_price}"

            print(f"{item[0]} X {item[1]}{'-' * (50 - len(element))}${new_price}")

    def find_price(self, item):
        self.temp_price = f"{(item[0] * item[2]):.2f}"


class SumCar(AddCar):
    car_total_price = 0

    def __init__(self, price_, amount_):
        self.price = price_
        self.amount = amount_

        SumCar.car_total_price += self.price * self.amount
        AddCar.car_elements += amount_


product = ""

while product.lower() != "p":
    if AddCar.car_elements != 0:
        product = input("¿Qué mas llevará?\n"
                        "presione P para pagar\n")

    else:
        product = input("Introdzca el producto que necesita\n"
                        "presione P para pagar\n")

    if product.lower() != "p":
        try:
            price = products[product.capitalize()]

            amount = int(input("Cantidad que desea comprar\n"))

            register = SumCar(price, amount)

            register.add_ticket()

            print(f"Cantidad de productos {SumCar.car_elements}\n\n"
                  f"Total : {register.car_total_price : .2f}\n\n")

        except KeyError:
            print(f"\n{product} Sin Existencias")

    else:
        print(f"\nPAGANDO... {SumCar.car_total_price : .2f}")
```
