```python
while True:
    # Selección del usuario sobre la operación que desea realizar
    
    selection = input("""Que operacion desea hacer:
    Suma
    Resta
    Multiplicación
    División
    Potencia
    """)
    
    if selection.lower() in ["suma", "resta", "multiplicacion", "multiplicación", "división", "division", "potencia"]:
        nums_input = ""

        while True:
            # Entrada de los números en (lista, set o diccionario) que desea operar el usuario
            
            nums_input = input("Ingrese dos nuemeros para sumar (recuerde usar: lista, set o diccionario)\n")

            if ("{" and "}") in nums_input or ("[" and "]") in nums_input or '"' in nums_input:
                break

            print("ERROR: recuerde usar: lista, set o diccionario")

        # Se transforma la seleccion de operador del usuario a un operador en cadena
        
        if selection == "suma":
            selection = "+"

        elif selection == "resta":
            selection = "-"

        elif selection == "multiplicacion" or selection == "multiplicación":
            selection = "*"

        elif selection == "division" or selection == "división":
            selection = "/"

        elif selection == "potencia":
            selection = "**"

        operator = f'{selection}='
        
        # Se encuentra el tipo de entrada que dió el usuario, para operar (lista, set o diccionario)
        
        try:
            nums = eval(nums_input)

            break

        except NameError:
            try:
                if ("[" or "]") not in nums_input:
                    nums = set(nums_input)

                else:
                    nums = list(nums_input)

                break

            except NameError:
                pass

    else:
        # Saltará este error cuando el usuario introduzca un operador erroneo
        
        print("Seleccione una operación correcta")


        
# Al salir de while se llevará a cabo la operacion        
result = 0

if type(nums) is dict:
    # Diccionario
    
    for number in nums:
        if not result and type(nums[number]) is not str:
            result = nums[number]

        elif type(nums[number]) is not str:
            result = f"{result} {selection} {nums[number]}"

            result = eval(result)

elif type(nums) is set:
    # set
    
    nums_input = nums_input.replace("{", "").replace("}", "").split(",")
    for number in nums_input:
        try:
            number = eval(number)

            if not result and type(number) is not str:
                result = number

            elif type(number) is not str:
                result = f"{result} {selection} {number}"

                result = eval(result)

        except NameError:
            pass

elif type(nums) is list:
    # lista
    
    nums_input = nums_input.replace("[", "").replace("]", "").split(",")
    for number in nums_input:
        try:
            number = eval(number)

            if not result and type(number) is not str:
                result = number

            elif type(number) is not str:
                result = f"{result} {selection} {number}"

                result = eval(result)

        except (SyntaxError, NameError):
            pass

        # Resultado de la operación

print("El resultado es: ", result)
```
